#Method 1

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
```

```python
OLLAMA_API = "http://localhost:11434/api/chat"
HEADERS = {"Content-Type": "application/json"}
MODEL = "llama3.2"
```

```python
messages = [{
    "role": "user", "content": "Write any small random recursive function to pick my brain"
}]
```

```python
payload = {
    "model": MODEL,
    "messages": messages,
    "stream": False
}
```

```python
response = requests.post(OLLAMA_API, headers=HEADERS, json=payload)
print(response.json()['message']['content'])
```

#Method 2

```python
import ollama

messages = [{
    "role":"user", "content":"Is it rude to video call someone who is kindof an introvert and someone who i have known for 7 months and she is a girl?"
}]

response = ollama.chat(MODEL, messages)
print(response['message']['content'])
```

#Method 3

```python
from openai import OpenAI

openai = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

response = openai.chat.completions.create(
    model = "llama3.2",
    messages = [{
    "role": "user",
    "content": "Why after POSH training, I am now afraid to talk to women?"
}])

print(response.choices[0].message.content)
```
